# Unsupervised training exhample (clustering)

https://scikit-learn.org/dev/api/sklearn.datasets.html

```python
from sklearn.datasets import make_blobs # creating synthetic dataset
```

```python
X, y = make_blobs(n_samples=10000, centers=10, cluster_std=0.80) # centers - how many blobs there will be, standard deviation cluster_std - the smaller the number (0.1) the denser and smaller the blobs are
```

```python
X[:, 0] # x coordinate
X[:, 1] # y coordinate
y # the cluster of the point
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c='grey', edgecolors='k') # c=y defines the color of each blob by the y value, c='grey' makes all data look the same
```

```python
# import model
from sklearn.cluster import KMeans

# initialize model
kmeans = KMeans(n_clusters=5) # n_clusters - how many clusters there will be

# fit the model
kmeans.fit(X) # training the model

# the prediction - call the trained model
# the outcome of the model is the cluster labels
cluster_labels = kmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='rainbow', edgecolor='b')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, marker='*', c='red', edgecolor='k') # adding cluster center to the cluster

```
