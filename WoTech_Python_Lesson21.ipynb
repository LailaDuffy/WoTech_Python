```python
import numpy as np

# Random coin toss

coin = ["heads", "tails"]
throws = 1000
throws_1000 = np.random.choice(coin, size=throws, p=[0.5, 0.5])

total_heads = np.sum(throws_1000 == "heads")
total_tails = np.sum(throws_1000 == "tails")

print("Total Heads:", total_heads)
print("Total Tails:", total_tails)
```

```python
# Coin toss with defined probablility

throws_1000_biased = np.random.choice(coin, size=throws, p=[0.3, 0.7])

total_heads_biased = np.sum(throws_1000_biased == "heads")
total_tails_biased = np.sum(throws_1000_biased == "tails")

print("Total Heads:", total_heads_biased)
print("Total Tails:", total_tails_biased)
```

```python
# Coin landin gon a side

coin_toss_side = ["heads", "tails", "side"]
throws_10000 = 10000

throws_10000 = np.random.choice(coin_toss_side, size=throws, p=[0.49165, 0.49165, 0.0167])

total_heads_side = np.sum(throws_10000 == "heads")
total_tails_side = np.sum(throws_10000 == "tails")
total_side = np.sum(throws_10000 == "side")

print("Total Heads:", total_heads_side)
print("Total Tails:", total_tails_side)
print("Total Side:", total_side)
```

```python
# Bar chart
import matplotlib.pyplot as plt

plt.bar(coin_toss_side, [total_heads_side, total_tails_side, total_side], color="teal")
plt.title("Coin Toss with Side landing")
plt.xlabel("Landing result")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar chart using np.unique

unique, counts = np.unique(throws_10000, return_counts=True)
plt.bar(unique, counts, color="teal")
plt.title("Coin Toss with Side landing")
plt.xlabel("Landing result")
plt.ylabel("Frequency")
plt.show()
```
